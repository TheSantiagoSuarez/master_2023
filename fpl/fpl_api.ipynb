### Player Database

```python
import urllib.request, json 
import pandas as pd
import numpy as np

with urllib.request.urlopen("https://draft.premierleague.com/api/bootstrap-static") as url:
    data = json.load(url)
    print(data)
```

```python
#get all player ids
df = pd.DataFrame(columns=["player_id","player_name","team_id","position"])

for i in data["elements"]:
    df = df.append({"player_id":i["id"], "player_name":i["web_name"], "team_id":i["team"], "position":i["element_type"]},ignore_index=True)
```

```python
#get all team ids
teams = pd.DataFrame(columns=["team_id","team"])

for i in data["teams"]:
    teams = teams.append({"team_id":i["id"],"team":i["short_name"]},ignore_index=True)
```

```python
#merging team names with player df
players = pd.merge(df, teams, on='team_id')
players = players.drop(columns="team_id")
players
```


### Lined Up Players Database

```python
managers = [893488,891742,893526,900304,896325,897752,900183,899952]
current_gw = 36
all_lineups_ser = pd.Series()

for gw in range(5,current_gw):
    s = pd.Series()
    for manager in managers:
        with urllib.request.urlopen(f"https://draft.premierleague.com/api/entry/{manager}/event/{gw}") as url:
            data = {manager:json.load(url)}
            s = s.append(pd.Series(data))
    all_lineups_ser = all_lineups_ser.append(pd.Series({gw:s}))    
```

```python
#get all manager picks

picks = pd.DataFrame(columns=["player_id","squad_position","gameweek","manager_id"])

for gw in range(5,current_gw):
    for manager in managers:
        for i in all_lineups_ser[gw][manager]["picks"]:
            picks = picks.append({"player_id":i["element"], "squad_position":i["position"], "gameweek":gw, "manager_id":manager},ignore_index=True)
```

### Points Scored and Stats Database

```python
player_stats_ser = pd.Series()

for gw in range(5,current_gw):
    with urllib.request.urlopen(f"https://draft.premierleague.com/api/event/{gw}/live") as url:
        data = {gw:json.load(url)}
    player_stats_ser = player_stats_ser.append(pd.Series(data)) 
```

```python
#function to add points. It checks if player exists, if it has a double gameweek or if it has a blank gameweek
def get_points(gw,player_id):
    points = 0

    try:
        if player_stats_ser[gw]["elements"][str(player_id)]:
            if not player_stats_ser[gw]["elements"][str(player_id)]["explain"]:
                print(f"Blank for player {player_id} in gameweek {gw}")
            else:
                for i in player_stats_ser[gw]["elements"][str(player_id)]['explain'][0][0]:
                    points += i["points"]
                if len(player_stats_ser[gw]["elements"][str(player_id)]["explain"])==2:
                    for i in player_stats_ser[gw]["elements"][str(player_id)]['explain'][1][0]:
                        points += i["points"]             
    except KeyError:
        print(f"No data for player {player_id} in gameweek {gw}")
        return 0
    
    return points
```

```python
#get all player stats per gameweek

player_stats = pd.DataFrame()

for gw in range(5,current_gw):
    for player_id in players["player_id"]:
        next_row = len(player_stats) + 1
        print(player_id)
        player_stats.loc[next_row,"player_id"] = player_id
        player_stats.loc[next_row,"gameweek"] = gw
        player_stats.loc[next_row,"points"] = get_points(gw,player_id)

        try:
            for key,value in player_stats_ser[gw]["elements"][str(player_id)]["stats"].items():
                player_stats.loc[next_row,key] = value
        except KeyError:
            continue
```

Merging the datasets

```python
x = pd.merge(player_stats,picks,how="left",on=["player_id","gameweek"])
akoya_df = pd.merge(x,players,how="left",on=["player_id"])


akoya_df["manager_id"] = akoya_df["manager_id"].fillna("transfer market")
akoya_df["manager_id"] = akoya_df["manager_id"].replace({893488:"Ali",891742:"Ruslan",
                                                         893526:"Sami",900304:"Yahya",
                                                         896325:"Youssef",897752:"Santi",
                                                         900183:"Shrey",899952:"Dani"})
```

```python
akoya_df["bench"] = np.where(akoya_df["squad_position"]>11, 'bench', np.where(akoya_df["squad_position"]>=1, 'fielded', 'n/a'))
akoya_df["position"] = akoya_df["position"].replace({1:"GK",2:"DEF",
                                                         3:"MID",4:"FWD"})
```

```python
akoya_df = akoya_df.reindex(columns=['player_id', 'player_name','position', 'team', 'gameweek', 
        'manager_id', "squad_position", 'bench', 'points', 'minutes', 'goals_scored', 'assists',
        'clean_sheets', 'goals_conceded', 'own_goals', 'penalties_saved',
        'penalties_missed', 'yellow_cards', 'red_cards', 'saves', 'bonus',
        'bps', 'influence', 'creativity', 'threat', 'ict_index', 'total_points','in_dreamteam'])
```
