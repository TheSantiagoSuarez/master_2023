**Explanatory Data Analysis - P03 2nd Half**

*Importing the Required libraries*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

*Fetching Data in the Directory for P03*

```python
# Directory where the CSV files are located
directory_p03_2ndhalf = "C:/Users/Jorge/Desktop/IE/3rd_Term/Capstone/Processed_Data/P03_2ndHalf"

# Pattern to match CSV files
file_pattern = "*.csv"

# Get the file names matching the pattern for P03
file_names_p03 = glob.glob(directory_p03_2ndhalf + "/" + file_pattern)
```

***For Loop** for **Exploratory Data Analysis** for **P03_2nd Half**, **Plotting**, **Statistics** and **Data Summary***

```python
for file_name in file_names_p03:
    # Read the CSV file
    df_p03 = pd.read_csv(file_name, delimiter=";")

    # Loop to replace "," in y-variable column and convert into float
    for column_name in df_p03.columns:
        if df_p03[column_name].dtype == object and df_p03[column_name].str.contains(",").any():
            df_p03[column_name] = df_p03[column_name].str.replace(",", ".").astype(float)

    # Convert time columns into datetime
    for col_name in df_p03.columns:
        if "time" in col_name.lower():
            df_p03[col_name] = pd.to_datetime(df_p03[col_name].str.replace(" CET", ""))

    # Filter columns with float data type
    float_columns_p03 = df_p03.select_dtypes(include=[np.float])

    # Plot columns with float data type for P03
    for column_name in float_columns_p03.columns:
        data_p03 = float_columns_p03[column_name]

        # Calculate statistics for P03
        minimum_p03 = np.min(data_p03)
        maximum_p03 = np.max(data_p03)
        average_p03 = np.mean(data_p03)
        std_dev_p03 = np.std(data_p03)
        outliers_p03 = data_p03[(data_p03 < average_p03 - 3 * std_dev_p03) | (data_p03 > average_p03 + 3 * std_dev_p03)]  # 3 z-scores from the mean
        num_outliers_p03 = len(outliers_p03)

        # Line plot using Matplotlib for P03
        fig, ax = plt.subplots()
        ax.plot(df_p03["time"], data_p03)
        ax.set_title("Line Plot (P03)")
        ax.set_xlabel("Time")
        ax.set_ylabel(column_name)

        # Add statistics to the plot for P03
        stats_text = f"Min: {minimum_p03:.2f}\nMax: {maximum_p03:.2f}\nAvg: {average_p03:.2f}\nOutliers: {num_outliers_p03}"
        ax.text(0.95, 0.9, stats_text, transform=ax.transAxes, ha="right", va="top",
                bbox=dict(facecolor="white", edgecolor="gray"))

        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # Box plot using Matplotlib for P03
        fig_box, ax_box = plt.subplots()
        ax_box.boxplot(data_p03)
        ax_box.set_title("Box Plot (P03)")
        ax_box.set_xlabel(column_name)
        ax_box.set_ylabel("Value")

        # Add statistics to the plot for P03
        stats_text = f"Min: {minimum_p03:.2f}\nMax: {maximum_p03:.2f}\nAvg: {average_p03:.2f}\nOutliers: {num_outliers_p03}"
        ax_box.text(1, maximum_p03, stats_text, ha="left", va="top", bbox=dict(facecolor="white", edgecolor="gray"))

        plt.tight_layout()
        plt.show()
```
